```python
import pandas as pd
import glob
import json
import re
```

```python
def return_useful_data(r_out):
    git_hits = 0
    osf_hits = 0
    nda_hits = 0
    open_neuro = 0
    ndar_hits = 0
    ndar_long_hits = 0
    #fcon1000_hits = 0 this is the same as indi?
    allen_institute_hits = 0
    hcp_hits = 0
    balsa_hits = 0
    loni_hits = 0
    loni_narrow_hits = 0
    fmridc_hits = 0
    ccrns_hits = 0
    datalad_hits = 0
    dataverse_hits = 0
    dbgap_hits = 0
    dbgap_url_hits = 0
    dryad_hits = 0
    figshare_hits = 0
    indi_hits = 0
    nitrc_hits = 0
    omega_hits = 0
    xnat_hits = 0
    zenodo_hits = 0
    aws_data_hits = 0

    try:
        doi = r_out['documents'][0]['passages'][0]['infons']['article-id_doi']
    except:
        doi = None
    try:
        pmcid = r_out['documents'][0]['id']
    except:
        pmcid = None
    for passage in r_out['documents'][0]['passages']:
        if re.search(re.compile('github'), passage['text'].lower()):
            git_hits = git_hits + 1
        if re.search(re.compile('osf\.io'), passage['text'].lower()):
            osf_hits = osf_hits + 1
        if re.search(re.compile('nda\.nih\.gov'), passage['text'].lower()):
            nda_hits = nda_hits + 1
        if re.search(re.compile('openneuro'), passage['text'].lower()):
            open_neuro = open_neuro + 1
        if re.search(re.compile('ndar'), passage['text'].lower()):
            ndar_hits = ndar_hits + 1    
        if re.search(re.compile('national database for autism research'), passage['text'].lower()):
            ndar_long_hits = ndar_long_hits + 1 
        if re.search(re.compile('brain-map\.org'), passage['text'].lower()):
            allen_institute_hits = allen_institute_hits + 1 
        if re.search(re.compile('humanconnectome\.org'), passage['text'].lower()):
            hcp_hits = hcp_hits + 1 
        if re.search(re.compile('balsa\.wustl\.edu'), passage['text'].lower()):
            balsa_hits = balsa_hits + 1 
        if re.search(re.compile('loni\.usc\.edu'), passage['text'].lower()):
            loni_hits = loni_hits + 1
        if re.search(re.compile('ida\.loni\.usc\.edu'), passage['text'].lower()):
            loni_narrow_hits = loni_narrow_hits + 1
        if re.search(re.compile('fmridc'), passage['text'].lower()):
            fmridc_hits = fmridc_hits + 1
        if re.search(re.compile('ccrns'), passage['text'].lower()):
            ccrns_hits = ccrns_hits + 1
        if re.search(re.compile('datalad'), passage['text'].lower()):
            datalad_hits = datalad_hits + 1
        if re.search(re.compile('dataverse'), passage['text'].lower()):
            dataverse_hits = dataverse_hits + 1
        if re.search(re.compile('dbgap'), passage['text'].lower()):
            dbgap_hits = dbgap_hits + 1
        if re.search(re.compile('nih\.gov\/gap'), passage['text'].lower()):
            dbgap_url_hits = dbgap_url_hits + 1
        if re.search(re.compile('dryad'), passage['text'].lower()):
            dryad_hits = dryad_hits + 1
        if re.search(re.compile('figshare'), passage['text'].lower()):
            figshare_hits = figshare_hits + 1
        if re.search(re.compile('fcon_1000\.projects'), passage['text'].lower()):
            indi_hits = indi_hits + 1
        if re.search(re.compile('nitrc'), passage['text'].lower()):
            nitrc_hits = nitrc_hits + 1
        if re.search(re.compile('mcgill\.ca\/bic\/resources\/omega'), passage['text'].lower()):
            omega_hits = omega_hits + 1
        if re.search(re.compile('xnat\.org'), passage['text'].lower()):
            xnat_hits = xnat_hits + 1
        if re.search(re.compile('zenodo'), passage['text'].lower()):
            zenodo_hits = zenodo_hits + 1
        if re.search(re.compile('opendata\.aws'), passage['text'].lower()):
            aws_data_hits = aws_data_hits + 1
        
    return(doi, pmcid, git_hits, osf_hits, nda_hits, open_neuro, ndar_hits, 
           ndar_long_hits, allen_institute_hits, hcp_hits, balsa_hits, loni_hits,
          loni_narrow_hits, fmridc_hits, ccrns_hits, datalad_hits, dataverse_hits,
          dbgap_hits, dryad_hits, figshare_hits, indi_hits, nitrc_hits, omega_hits,
          xnat_hits, zenodo_hits, aws_data_hits)
```

```python
already_retrieved = glob.glob('/home/riddleta/ac_knowl/output/full_texts/papes*')
already_retrieved = sorted(already_retrieved)
status_prints = already_retrieved[::2]
```

```python
data_collect = []

for i in already_retrieved:
    if i in status_prints:
        print(i)
    with open(i) as infile:
        dat = json.load(infile)
        for j, paper in enumerate(dat):
            data_collect.append(return_useful_data(paper))    
```

```python
df = pd.DataFrame(data_collect, columns = ['doi', 'pmcid', 'git_hits', 'osf_hits', 'nda_hits',
                                          'open_neuro', 'ndar_hits', 'ndar_long_hits', 'allen_institute_hits',
                                          'hcp_hits', 'balsa_hits', 'loni_hits', 'loni_narrow_hits', 'fmridc_hits',
                                          'ccrns_hits', 'datalad_hits', 'dataverse_hits', 'dbgap_hits',
                                          'dryad_hits', 'figshare_hits', 'indi_hits', 'nitrc_hits', 'omega_hits',
                                          'xnat_hits', 'zenodo_hits', 'aws_data_hits'])
```
